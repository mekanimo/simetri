```python
import simetri.graphics as sg

canvas = sg.Canvas()

points  = [(0, 0), (200, 0)]
kernel = sg.Shape(points)
transform = sg.rotation_matrix(sg.pi/24) @ sg.scale_matrix(.97, .97)
pattern = kernel.transform(transform, reps=47)
pattern = sg.frieze.spinning_hop(pattern, (130, 0), 240, 140, 3)
pattern.rotate(sg.pi/8, (-400, -400), reps=15).scale(.25)
canvas.draw(pattern, line_width=2, line_cap = sg.LineCap.ROUND, line_color=sg.navy)

canvas.display()
```

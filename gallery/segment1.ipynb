```python
import simetri.graphics as sg

canvas = sg.Canvas()

points  = [(0, 0), (200, 0)]
kernel = sg.Shape(points)
transform = sg.rotation_matrix(sg.pi/24) @ sg.scale_matrix(.97, .97)
pattern = kernel.transform(transform, reps=47)
pattern = sg.frieze.spinning_hop(pattern, (130, 0), 180, 180, 3).scale(.5)
canvas.draw(pattern, line_width=3, line_cap = sg.LineCap.ROUND, line_color=sg.navy)

canvas.display()
```
